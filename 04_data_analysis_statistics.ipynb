  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">Motivation — Why Do We Need Statistics?</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">Imagine a population of 10 observations made up of four categories: <code>A</code>, <code>B</code>, <code>C</code>, <code>D</code>. The population is: <strong>{B, A, B, D, B, C, A, B, D, B}</strong>. In the real world we rarely see the <em>whole</em> population — we only see a <em>sample</em>, and we must use it to answer questions such as:</p>
  <ul style="color:#333; line-height:1.8;">
    <li>Is the proportion of <strong>C</strong> and <strong>D</strong> the same?</li>
    <li>Is the proportion of <strong>C</strong> equal to 40%? What about 10%?</li>
    <li>Is the proportion of <strong>B</strong> greater than 60%?</li>
  </ul>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">These are <em>inference</em> questions — we answer them with the tools of descriptive and inferential statistics.</p>
</div>

````python
# Visualise the population of N = 10 and a sample of n = 5
population = np.array(['B', 'A', 'B', 'D', 'B', 'C', 'A', 'B', 'D', 'B'])
sample = np.array(['A', 'B', 'B', 'C', 'D'])

pop_counts = pd.Series(population).value_counts().reindex(['A', 'B', 'C', 'D'])
sam_counts = pd.Series(sample).value_counts().reindex(['A', 'B', 'C', 'D']).fillna(0)

pop_rel = pop_counts / pop_counts.sum()
sam_rel = sam_counts / sam_counts.sum()

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].bar(pop_rel.index, pop_rel.values, color='#4B2E83', edgecolor='white')
axes[0].set_title('Population Relative Frequency (N = 10)', fontsize=13, color='#4B2E83', fontweight='bold')
axes[0].set_xlabel('Category')
axes[0].set_ylabel('Relative Frequency')
axes[0].set_ylim(0, 0.7)
for i, v in enumerate(pop_rel.values):
    axes[0].text(i, v + 0.01, f'{v:.2f}', ha='center', color='#333')

axes[1].bar(sam_rel.index, sam_rel.values, color='#E8B923', edgecolor='white')
axes[1].set_title('Sample Relative Frequency (n = 5)', fontsize=13, color='#4B2E83', fontweight='bold')
axes[1].set_xlabel('Category')
axes[1].set_ylabel('Relative Frequency')
axes[1].set_ylim(0, 0.7)
for i, v in enumerate(sam_rel.values):
    axes[1].text(i, v + 0.01, f'{v:.2f}', ha='center', color='#333')

fig.text(0.5, -0.02, 'Source: Illustrative population and Sample 1 (n = 5)', ha='center', color='#666', fontsize=10)
plt.tight_layout()
plt.show()
````

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">Definitions: Population &amp; Sample</h2>
</div>

| Concept | Definition | Notation | Example |
|:--------|:-----------|:---------|:--------|
| **Population** | The complete set of all individuals, objects, or measurements of interest in a study. | **N** (population size) | All registered voters in a country; every product from a factory. |
| **Sample** | A subset of the population selected for observation and analysis. | **n** (sample size, n ≤ N) | 1,200 voters polled; 50 items inspected. |

**Key requirement:** the sample must be *representative* of the population. Random sampling reduces bias, and larger *n* produces more precise estimates.

> **Why sample?** Studying the entire population is often impossible, impractical, or too costly. Instead, we draw a sample and use it to make inferences about the population. This is the foundation of **statistical inference**.

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">Why Descriptive Statistics?</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">Raw data is overwhelming. A population of even five values is hard to grasp at a glance — imagine thousands or millions of observations. Descriptive statistics give us tools to <strong>summarise</strong> and <strong>visualise</strong> the data.</p>
</div>

| Family | Statistic | Meaning |
|:-------|:----------|:--------|
| **Center** | Mean (μ) | Balancing point of the data |
| | Median | Middle value — robust to outliers |
| | Mode | Most frequent value |
| **Spread** | Variance / Std Dev (σ², σ) | Average squared distance from the mean |
| | Range | max − min |
| | IQR | Q3 − Q1 — robust spread of the middle 50% |
| **Position / Extremes** | Min, Q1, Q3, Max | Five-number summary |

> *"A picture is worth a thousand numbers."* Plots such as histograms, box plots, and scatter plots reveal shapes, skewness, clusters, gaps, and outliers that summary statistics alone cannot convey.

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">Worked Example: Population {3, 4, 2, 1, 5}</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">A population of size N = 5. Let us compute every descriptive statistic discussed above.</p>
</div>

````python
pop = np.array([3, 4, 2, 1, 5])

mean = pop.mean()
median = np.median(pop)
var_pop = pop.var(ddof=0)          # population variance: divide by N
std_pop = pop.std(ddof=0)
rng = pop.max() - pop.min()
q1, q3 = np.percentile(pop, [25, 75])
iqr = q3 - q1

print(f'Population = {pop.tolist()}')
print(f'Mean (\u03bc)         = {mean:.3f}')
print(f'Median           = {median:.3f}')
print(f'Variance (\u03c3\u00b2)    = {var_pop:.3f}')
print(f'Std Dev (\u03c3)      = {std_pop:.3f}')
print(f'Range            = {rng}')
print(f'Q1, Q3           = {q1}, {q3}')
print(f'IQR              = {iqr}')
print(f'Min, Max         = {pop.min()}, {pop.max()}')
````

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">The Mean as Center of Mass</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">The mean (μ) is the arithmetic average — it is the <em>balance point</em> of equal weights placed on a number line. For the population above: μ = (3 + 4 + 2 + 1 + 5) / 5 = 15 / 5 = <strong>3.0</strong>.</p>
</div>

````python
pop = np.array([3, 4, 2, 1, 5])
mu = pop.mean()

fig, ax = plt.subplots(figsize=(10, 4))

# Number line
ax.axhline(0, color='#333', linewidth=1.5)

# Plot each data point as a filled circle
for i, x in enumerate(pop):
    ax.plot(x, 0, 'o', markersize=18, color='#4B2E83', markeredgecolor='white', markeredgewidth=2)
    ax.text(x, 0.12, f'x{i+1}={x}', ha='center', color='#4B2E83', fontweight='bold')

# Fulcrum at the mean
ax.plot(mu, -0.15, '^', markersize=25, color='#E8B923', markeredgecolor='#4B2E83', markeredgewidth=2)
ax.text(mu, -0.28, f'\u03bc = {mu:.1f}', ha='center', color='#E8B923', fontweight='bold', fontsize=13)

ax.set_xlim(0, 6)
ax.set_ylim(-0.4, 0.4)
ax.set_yticks([])
ax.set_xlabel('Value', fontsize=12, color='#333')
ax.set_title('The Mean is the Balance Point (Center of Mass)', fontsize=14, color='#4B2E83', fontweight='bold')
for spine in ['top', 'right', 'left']:
    ax.spines[spine].set_visible(False)

fig.text(0.5, -0.05, 'Source: Population {3, 4, 2, 1, 5}', ha='center', color='#666', fontsize=10)
plt.tight_layout()
plt.show()
````

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">Variance: Spread Around the Mean</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">The variance (σ²) measures how far values spread from the mean. We compute the squared deviations, average them, and take the square root to recover the standard deviation (σ).</p>
</div>

$$\sigma^2 \;=\; \frac{1}{N} \sum_{i=1}^{N} (x_i - \mu)^2$$

| xᵢ | xᵢ − μ | (xᵢ − μ)² |
|:--:|:------:|:---------:|
| 3  | 0      | 0         |
| 4  | +1     | 1         |
| 2  | −1     | 1         |
| 1  | −2     | 4         |
| 5  | +2     | 4         |

σ² = (0 + 1 + 1 + 4 + 4) / 5 = 10 / 5 = **2.0**   →   σ = √2 ≈ **1.414**

````python
pop = np.array([3, 4, 2, 1, 5])
mu = pop.mean()
dev = pop - mu
sq_dev = dev ** 2
var_pop = sq_dev.mean()

print('Squared deviations from the mean:')
for x, d, s in zip(pop, dev, sq_dev):
    print(f'  x = {x}  |  x - \u03bc = {d:+.0f}  |  (x - \u03bc)\u00b2 = {s:.0f}')

print(f'\nVariance (\u03c3\u00b2) = {var_pop:.3f}')
print(f'Std Dev  (\u03c3)  = {np.sqrt(var_pop):.3f}')
````

````python
scores = np.array([72, 85, 90, 68, 95, 78, 88, 84])

# Calculate mean and median
mean_scores = np.mean(scores)
median_scores = np.median(scores)

# Calculate range
range_scores = np.max(scores) - np.min(scores)

# Calculate population variance and standard deviation (ddof=0 for population)
var_scores = np.var(scores, ddof=0)
std_scores = np.std(scores, ddof=0)

# Calculate IQR (Q3 - Q1)
q1_scores, q3_scores = np.percentile(scores, [25, 75])
iqr_scores = q3_scores - q1_scores

# Print results
print(f'Exam Scores Population: {scores.tolist()}')
print(f'Population Mean (μ): {mean_scores:.3f}')
print(f'Median: {median_scores:.3f}')
print(f'Range: {range_scores}')
print(f'Population Variance (σ²): {var_scores:.3f}')
print(f'Population Std Dev (σ): {std_scores:.3f}')
print(f'IQR (Q3 - Q1): {iqr_scores:.3f}')
````

*Write your written answer (ddof meaning, IQR, symmetry) in this cell.*

1. **`ddof=0`** — `ddof` stands for "Delta Degrees of Freedom". When calculating variance or standard deviation, `ddof` determines the divisor used in the calculation. Setting `ddof=0` means the denominator used is `N` (the total number of observations). This is used when you are treating your data as the *entire population* and want to calculate the population variance (σ²) or population standard deviation (σ).
2. **IQR computation** —In `numpy`, the Interquartile Range (IQR) is typically computed by finding the 25th percentile (Q1) and the 75th percentile (Q3) of the data, and then subtracting Q1 from Q3. The `np.percentile(scores, [25, 75])` function directly returns these two values, and their difference gives the IQR.

3.  **Mean vs median symmetry** — For the exam scores, the mean (μ) is 82.500 and the median is 84.500. Since the mean is slightly less than the median, this suggests that the distribution of exam scores is **slightly left-skewed (negatively skewed)**. In a perfectly symmetric distribution, the mean and median would be equal.

**Context:** Two factories produce bolts targeting a length of **50.0 mm**. Factory A makes precision bolts; Factory B makes general bolts. Compare their variability.

**Data:**

- Factory A (N = 6): `A = [50.1, 49.8, 50.3, 50.0, 49.9, 50.2]`
- Factory B (N = 6): `B = [48.5, 51.2, 49.0, 51.8, 50.5, 49.3]`

💬 **LLM Prompt**

```
Using the Factory A and Factory B bolt measurements above, write Python code that for each factory computes:
- Mean (μ), variance (σ²), std dev (σ)
- Coefficient of Variation: CV = (σ/μ) × 100%
Print a side-by-side comparison table.
Use numpy with population variance (ddof=0).
```

🎯 **Task:** Run the code. Then explain: (1) Why is CV more useful than σ alone for comparing variability? (2) Which factory is more consistent and how does the code show this? (3) Why does quality control care about CV?

````python
A = np.array([50.1, 49.8, 50.3, 50.0, 49.9, 50.2])
B = np.array([48.5, 51.2, 49.0, 51.8, 50.5, 49.3])

# --- Your code here ---

def calculate_stats(data, name):
    mean = np.mean(data)
    var = np.var(data, ddof=0)
    std = np.std(data, ddof=0)
    cv = (std / mean) * 100
    return {'Factory': name, 'Mean (μ)': mean, 'Variance (σ²)': var, 'Std Dev (σ)': std, 'CV (%)': cv}

stats_A = calculate_stats(A, 'Factory A')
stats_B = calculate_stats(B, 'Factory B')

# Create a DataFrame for side-by-side comparison
comparison_df = pd.DataFrame([stats_A, stats_B]).set_index('Factory')

print("Bolt Measurements Comparison:")
display(comparison_df.round(3))
````

*Write your written answer here.*

Why is CV more useful than σ alone for comparing variability? The standard deviation (σ) measures the absolute spread of data around the mean. However, comparing σ directly between datasets with very different means can be misleading. For example, a σ of 1 unit might be considered small for values around 1000, but very large for values around 10. The Coefficient of Variation (CV) expresses the standard deviation as a percentage of the mean (CV = (σ/μ) × 100%). This makes it a unitless measure of relative variability, allowing for a meaningful comparison of dispersion between datasets that have different means or are measured in different units.

Which factory is more consistent and how does the code show this? Factory A is more consistent. While both factories have the same mean bolt length (50.05 mm), Factory A has a significantly lower Standard Deviation (0.171 vs 1.201) and, more importantly, a much lower Coefficient of Variation (0.341% vs 2.400%). A lower CV indicates less relative variability around the mean, meaning Factory A produces bolts that are more consistently close to the target length compared to Factory B.

Why does quality control care about CV? **The central question of statistics:** How well does a statistic (calculated from the sample) estimate the true parameter of the population? The answer depends on *sampling variability* — different samples give different estimates — and we need tools to quantify this uncertainty.

What does scipy.stats.t.interval do step by step? scipy.stats.t.interval(confidence_level, df, loc=x_bar, scale=sem) calculates a confidence interval for the mean of a population, assuming the data follows a t-distribution. Here's a step-by-step breakdown:

confidence_level: This is the desired confidence level for the interval (e.g., 0.95 for a 95% CI).
df: This stands for 'degrees of freedom', which is n - 1 (where n is the sample size). It determines the shape of the t-distribution.
loc=x_bar: This sets the center of the interval to the sample mean (x̄).
scale=sem: This provides the standard error of the mean (SEM), which is s / sqrt(n), where s is the sample standard deviation and n is the sample size. The scale parameter essentially stretches or shrinks the interval based on the variability of the sample means.

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">What Is a Sampling Distribution?</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">A <strong>sampling distribution</strong> is the probability distribution of a statistic (e.g., x̄, p̂) obtained from <em>all possible</em> samples of size n drawn from a population.</p>
  <ul style="color:#333; line-height:1.8;">
    <li><strong>Step 1:</strong> Start with a population described by μ and σ.</li>
    <li><strong>Step 2:</strong> Draw many random samples (S₁, S₂, …, Sₖ) of the same size n.</li>
    <li><strong>Step 3:</strong> Compute the statistic for each sample (x̄₁, x̄₂, …, x̄ₖ) and plot the distribution.</li>
  </ul>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">The resulting distribution is centered at μ with spread σ/√n. This is the foundation of every confidence interval and hypothesis test you will ever compute.</p>
</div>

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">Sampling Variability &amp; Bias</h2>
</div>

**Sampling variability** is the natural variation in a statistic from sample to sample. It is *not* an error — it is an inherent property of sampling.

| Sources | Reducing Variability |
|:--|:--|
| Random selection — different individuals chosen | Increase *n* — SE = σ/√n shrinks |
| Population heterogeneity | Better sampling (stratification, clustering) |
| Small sample size | More precise measurements |

**Bias.** An estimator is *unbiased* if E(estimator) = parameter. Examples:
- E(x̄) = μ ✓
- E(p̂) = π ✓
- E(s²) = σ² (sample variance with df = n − 1) ✓

But using n (instead of n − 1) for the sample variance produces a **biased** estimator: E[Σ(xᵢ − x̄)² / n] = (n − 1)/n · σ² < σ². Non-random sampling (convenience, voluntary response, survivorship) also introduces bias.

> **MSE = Bias² + Variance.** A good estimator is both unbiased (centred on the target) and has low variance (tightly clustered).

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">History &amp; Intuition of the CLT</h2>
</div>

| Year | Contributor | Contribution |
|:----:|:------------|:-------------|
| 1733 | Abraham de Moivre | First observed that binomial distributions approximate the normal curve. |
| 1812 | Pierre-Simon Laplace | Generalised to sums of independent random variables. |
| 1920 | Lindeberg–Lévy | Proved the modern CLT under general conditions (i.i.d. with finite variance). |

**Modern formulation.** If X₁, …, Xₙ are i.i.d. with mean μ and finite variance σ², then:

$$\sqrt{n}\;\frac{\bar X - \mu}{\sigma} \;\longrightarrow\; N(0, 1) \quad\text{as } n \to \infty$$

**Three intuitions:**
1. **Cancellation of extremes** — averaging many values lets highs and lows cancel.
2. **Accumulation of information** — each observation adds a small piece; outliers lose influence.
3. **Probability concentration** — by the Law of Large Numbers, x̄ → μ; the CLT tells us the *shape* (normal) and *rate* (σ/√n) of that convergence.

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">Standard Error &amp; Confidence Intervals</h2>
</div>

**Standard Error (SE)** — the standard deviation of a *sampling distribution*. It quantifies how much a statistic varies from sample to sample.

| Statistic | Formula |
|:--|:--|
| Sample mean | SE = σ / √n |
| Sample proportion | SE = √( p̂(1 − p̂) / n ) |

Larger n → smaller SE → more precise estimates.

**Confidence Interval (CI)** — a range of values likely to contain the true parameter with a specified confidence level.

$$\text{95\% CI:}\quad \text{estimate} \;\pm\; z^* \cdot \text{SE}$$

**Interpretation:** if we repeated the sampling many times, 95% of the CIs would contain the true parameter.

> Error bars on a chart are **visual confidence intervals**. If the error bar for a sample proportion includes the hypothesised value, we cannot reject H₀ at that significance level — directly linking CIs to hypothesis testing.

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">Z vs Student's t</h2>
</div>

| | Normal (Z) | Student's t |
|:--|:--|:--|
| Formula | Z = (x̄ − μ) / (σ/√n) | t = (x̄ − μ) / (*s*/√n) |
| Use when | σ known, n ≥ 30 (CLT) | σ unknown, small n |
| Shape | Fixed symmetric bell curve | Heavier tails, depends on df = n − 1 |
| 95% critical value | z* = 1.96 | t* depends on df |

**95% critical values (two-sided):**

| df | 2 | 5 | 10 | 20 | 30 | 120 | ∞ (Z) |
|:--:|:-:|:-:|:-:|:-:|:-:|:-:|:-:|
| t* | 4.303 | 2.571 | 2.228 | 2.086 | 2.042 | 1.980 | **1.960** |

As df → ∞, t* → z*. For large samples the two distributions are indistinguishable.

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">Hypothesis Testing</h2>
</div>

| Symbol | Name | Role |
|:--|:--|:--|
| **H₀** | Null hypothesis | Default claim (always contains `=`). Assumed true until the data say otherwise. |
| **H₁** | Alternative hypothesis | The research claim (contains `≠`, `<`, `>`). Accepted only if H₀ is rejected. |
| **α** | Significance level | Probability of rejecting H₀ when it is actually true (Type I error). Typically 0.05. |
| **p-value** | | Probability of observing data as extreme as ours, assuming H₀ is true. |

**Decision rule:**

- If p-value < α → **reject H₀**
- If p-value ≥ α → **fail to reject H₀**

**CI ↔ Hypothesis test duality.** A (1 − α) confidence interval contains *all* values of the null hypothesis that would **not** be rejected by a two-sided test at level α:

- If μ₀ ∈ CI → fail to reject H₀
- If μ₀ ∉ CI → reject H₀

A 95% CI is therefore equivalent to a two-sided hypothesis test at α = 0.05.
