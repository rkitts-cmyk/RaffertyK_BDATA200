  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">Motivation — Why Do We Need Statistics?</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">Imagine a population of 10 observations made up of four categories: <code>A</code>, <code>B</code>, <code>C</code>, <code>D</code>. The population is: <strong>{B, A, B, D, B, C, A, B, D, B}</strong>. In the real world we rarely see the <em>whole</em> population — we only see a <em>sample</em>, and we must use it to answer questions such as:</p>
  <ul style="color:#333; line-height:1.8;">
    <li>Is the proportion of <strong>C</strong> and <strong>D</strong> the same?</li>
    <li>Is the proportion of <strong>C</strong> equal to 40%? What about 10%?</li>
    <li>Is the proportion of <strong>B</strong> greater than 60%?</li>
  </ul>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">These are <em>inference</em> questions — we answer them with the tools of descriptive and inferential statistics.</p>
</div>

````python
# Visualise the population of N = 10 and a sample of n = 5
population = np.array(['B', 'A', 'B', 'D', 'B', 'C', 'A', 'B', 'D', 'B'])
sample = np.array(['A', 'B', 'B', 'C', 'D'])

pop_counts = pd.Series(population).value_counts().reindex(['A', 'B', 'C', 'D'])
sam_counts = pd.Series(sample).value_counts().reindex(['A', 'B', 'C', 'D']).fillna(0)

pop_rel = pop_counts / pop_counts.sum()
sam_rel = sam_counts / sam_counts.sum()

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].bar(pop_rel.index, pop_rel.values, color='#4B2E83', edgecolor='white')
axes[0].set_title('Population Relative Frequency (N = 10)', fontsize=13, color='#4B2E83', fontweight='bold')
axes[0].set_xlabel('Category')
axes[0].set_ylabel('Relative Frequency')
axes[0].set_ylim(0, 0.7)
for i, v in enumerate(pop_rel.values):
    axes[0].text(i, v + 0.01, f'{v:.2f}', ha='center', color='#333')

axes[1].bar(sam_rel.index, sam_rel.values, color='#E8B923', edgecolor='white')
axes[1].set_title('Sample Relative Frequency (n = 5)', fontsize=13, color='#4B2E83', fontweight='bold')
axes[1].set_xlabel('Category')
axes[1].set_ylabel('Relative Frequency')
axes[1].set_ylim(0, 0.7)
for i, v in enumerate(sam_rel.values):
    axes[1].text(i, v + 0.01, f'{v:.2f}', ha='center', color='#333')

fig.text(0.5, -0.02, 'Source: Illustrative population and Sample 1 (n = 5)', ha='center', color='#666', fontsize=10)
plt.tight_layout()
plt.show()
````

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">Definitions: Population &amp; Sample</h2>
</div>

| Concept | Definition | Notation | Example |
|:--------|:-----------|:---------|:--------|
| **Population** | The complete set of all individuals, objects, or measurements of interest in a study. | **N** (population size) | All registered voters in a country; every product from a factory. |
| **Sample** | A subset of the population selected for observation and analysis. | **n** (sample size, n ≤ N) | 1,200 voters polled; 50 items inspected. |

**Key requirement:** the sample must be *representative* of the population. Random sampling reduces bias, and larger *n* produces more precise estimates.

> **Why sample?** Studying the entire population is often impossible, impractical, or too costly. Instead, we draw a sample and use it to make inferences about the population. This is the foundation of **statistical inference**.

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">Why Descriptive Statistics?</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">Raw data is overwhelming. A population of even five values is hard to grasp at a glance — imagine thousands or millions of observations. Descriptive statistics give us tools to <strong>summarise</strong> and <strong>visualise</strong> the data.</p>
</div>

| Family | Statistic | Meaning |
|:-------|:----------|:--------|
| **Center** | Mean (μ) | Balancing point of the data |
| | Median | Middle value — robust to outliers |
| | Mode | Most frequent value |
| **Spread** | Variance / Std Dev (σ², σ) | Average squared distance from the mean |
| | Range | max − min |
| | IQR | Q3 − Q1 — robust spread of the middle 50% |
| **Position / Extremes** | Min, Q1, Q3, Max | Five-number summary |

> *"A picture is worth a thousand numbers."* Plots such as histograms, box plots, and scatter plots reveal shapes, skewness, clusters, gaps, and outliers that summary statistics alone cannot convey.

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">Worked Example: Population {3, 4, 2, 1, 5}</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">A population of size N = 5. Let us compute every descriptive statistic discussed above.</p>
</div>

````python
pop = np.array([3, 4, 2, 1, 5])

mean = pop.mean()
median = np.median(pop)
var_pop = pop.var(ddof=0)          # population variance: divide by N
std_pop = pop.std(ddof=0)
rng = pop.max() - pop.min()
q1, q3 = np.percentile(pop, [25, 75])
iqr = q3 - q1

print(f'Population = {pop.tolist()}')
print(f'Mean (\u03bc)         = {mean:.3f}')
print(f'Median           = {median:.3f}')
print(f'Variance (\u03c3\u00b2)    = {var_pop:.3f}')
print(f'Std Dev (\u03c3)      = {std_pop:.3f}')
print(f'Range            = {rng}')
print(f'Q1, Q3           = {q1}, {q3}')
print(f'IQR              = {iqr}')
print(f'Min, Max         = {pop.min()}, {pop.max()}')
````

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">The Mean as Center of Mass</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">The mean (μ) is the arithmetic average — it is the <em>balance point</em> of equal weights placed on a number line. For the population above: μ = (3 + 4 + 2 + 1 + 5) / 5 = 15 / 5 = <strong>3.0</strong>.</p>
</div>

````python
pop = np.array([3, 4, 2, 1, 5])
mu = pop.mean()

fig, ax = plt.subplots(figsize=(10, 4))

# Number line
ax.axhline(0, color='#333', linewidth=1.5)

# Plot each data point as a filled circle
for i, x in enumerate(pop):
    ax.plot(x, 0, 'o', markersize=18, color='#4B2E83', markeredgecolor='white', markeredgewidth=2)
    ax.text(x, 0.12, f'x{i+1}={x}', ha='center', color='#4B2E83', fontweight='bold')

# Fulcrum at the mean
ax.plot(mu, -0.15, '^', markersize=25, color='#E8B923', markeredgecolor='#4B2E83', markeredgewidth=2)
ax.text(mu, -0.28, f'\u03bc = {mu:.1f}', ha='center', color='#E8B923', fontweight='bold', fontsize=13)

ax.set_xlim(0, 6)
ax.set_ylim(-0.4, 0.4)
ax.set_yticks([])
ax.set_xlabel('Value', fontsize=12, color='#333')
ax.set_title('The Mean is the Balance Point (Center of Mass)', fontsize=14, color='#4B2E83', fontweight='bold')
for spine in ['top', 'right', 'left']:
    ax.spines[spine].set_visible(False)

fig.text(0.5, -0.05, 'Source: Population {3, 4, 2, 1, 5}', ha='center', color='#666', fontsize=10)
plt.tight_layout()
plt.show()
````

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">Variance: Spread Around the Mean</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">The variance (σ²) measures how far values spread from the mean. We compute the squared deviations, average them, and take the square root to recover the standard deviation (σ).</p>
</div>

$$\sigma^2 \;=\; \frac{1}{N} \sum_{i=1}^{N} (x_i - \mu)^2$$

| xᵢ | xᵢ − μ | (xᵢ − μ)² |
|:--:|:------:|:---------:|
| 3  | 0      | 0         |
| 4  | +1     | 1         |
| 2  | −1     | 1         |
| 1  | −2     | 4         |
| 5  | +2     | 4         |

σ² = (0 + 1 + 1 + 4 + 4) / 5 = 10 / 5 = **2.0**   →   σ = √2 ≈ **1.414**

````python
pop = np.array([3, 4, 2, 1, 5])
mu = pop.mean()
dev = pop - mu
sq_dev = dev ** 2
var_pop = sq_dev.mean()

print('Squared deviations from the mean:')
for x, d, s in zip(pop, dev, sq_dev):
    print(f'  x = {x}  |  x - \u03bc = {d:+.0f}  |  (x - \u03bc)\u00b2 = {s:.0f}')

print(f'\nVariance (\u03c3\u00b2) = {var_pop:.3f}')
print(f'Std Dev  (\u03c3)  = {np.sqrt(var_pop):.3f}')
````

🎯 **Task:** Run the LLM-generated code. Then explain: (1) What does `ddof=0` mean and why use it here? (2) How does numpy compute the IQR? (3) What does comparing mean vs median tell you about symmetry?

````python
scores = np.array([72, 85, 90, 68, 95, 78, 88, 84])

# Calculate mean and median
mean_scores = np.mean(scores)
median_scores = np.median(scores)

# Calculate range
range_scores = np.max(scores) - np.min(scores)

# Calculate population variance and standard deviation (ddof=0 for population)
var_scores = np.var(scores, ddof=0)
std_scores = np.std(scores, ddof=0)

# Calculate IQR (Q3 - Q1)
q1_scores, q3_scores = np.percentile(scores, [25, 75])
iqr_scores = q3_scores - q1_scores

# Print results
print(f'Exam Scores Population: {scores.tolist()}')
print(f'Population Mean (μ): {mean_scores:.3f}')
print(f'Median: {median_scores:.3f}')
print(f'Range: {range_scores}')
print(f'Population Variance (σ²): {var_scores:.3f}')
print(f'Population Std Dev (σ): {std_scores:.3f}')
print(f'IQR (Q3 - Q1): {iqr_scores:.3f}')
````

*Write your written answer (ddof meaning, IQR, symmetry) in this cell.*

1. **`ddof=0`** — `ddof` stands for "Delta Degrees of Freedom". When calculating variance or standard deviation, `ddof` determines the divisor used in the calculation. Setting `ddof=0` means the denominator used is `N` (the total number of observations). This is used when you are treating your data as the *entire population* and want to calculate the population variance (σ²) or population standard deviation (σ).
2. **IQR computation** —In `numpy`, the Interquartile Range (IQR) is typically computed by finding the 25th percentile (Q1) and the 75th percentile (Q3) of the data, and then subtracting Q1 from Q3. The `np.percentile(scores, [25, 75])` function directly returns these two values, and their difference gives the IQR.

3.  **Mean vs median symmetry** — For the exam scores, the mean (μ) is 82.500 and the median is 84.500. Since the mean is slightly less than the median, this suggests that the distribution of exam scores is **slightly left-skewed (negatively skewed)**. In a perfectly symmetric distribution, the mean and median would be equal.

**Context:** Two factories produce bolts targeting a length of **50.0 mm**. Factory A makes precision bolts; Factory B makes general bolts. Compare their variability.

**Data:**

- Factory A (N = 6): `A = [50.1, 49.8, 50.3, 50.0, 49.9, 50.2]`
- Factory B (N = 6): `B = [48.5, 51.2, 49.0, 51.8, 50.5, 49.3]`

💬 **LLM Prompt**

```
Using the Factory A and Factory B bolt measurements above, write Python code that for each factory computes:
- Mean (μ), variance (σ²), std dev (σ)
- Coefficient of Variation: CV = (σ/μ) × 100%
Print a side-by-side comparison table.
Use numpy with population variance (ddof=0).
```

🎯 **Task:** Run the code. Then explain: (1) Why is CV more useful than σ alone for comparing variability? (2) Which factory is more consistent and how does the code show this? (3) Why does quality control care about CV?

````python
A = np.array([50.1, 49.8, 50.3, 50.0, 49.9, 50.2])
B = np.array([48.5, 51.2, 49.0, 51.8, 50.5, 49.3])

# --- Your code here ---

def calculate_stats(data, name):
    mean = np.mean(data)
    var = np.var(data, ddof=0)
    std = np.std(data, ddof=0)
    cv = (std / mean) * 100
    return {'Factory': name, 'Mean (μ)': mean, 'Variance (σ²)': var, 'Std Dev (σ)': std, 'CV (%)': cv}

stats_A = calculate_stats(A, 'Factory A')
stats_B = calculate_stats(B, 'Factory B')

# Create a DataFrame for side-by-side comparison
comparison_df = pd.DataFrame([stats_A, stats_B]).set_index('Factory')

print("Bolt Measurements Comparison:")
display(comparison_df.round(3))
````

*Write your written answer here.*

Why is CV more useful than σ alone for comparing variability? The standard deviation (σ) measures the absolute spread of data around the mean. However, comparing σ directly between datasets with very different means can be misleading. For example, a σ of 1 unit might be considered small for values around 1000, but very large for values around 10. The Coefficient of Variation (CV) expresses the standard deviation as a percentage of the mean (CV = (σ/μ) × 100%). This makes it a unitless measure of relative variability, allowing for a meaningful comparison of dispersion between datasets that have different means or are measured in different units.

| | **Parameter** (population) | **Statistic** (sample) |
|:--|:--|:--|
| Description | Fixed, unknown value describing the whole population | Calculated value that *estimates* the parameter |

**The central question of statistics:** How well does a statistic (calculated from the sample) estimate the true parameter of the population? The answer depends on *sampling variability* — different samples give different estimates — and we need tools to quantify this uncertainty.

What does scipy.stats.t.interval do step by step? scipy.stats.t.interval(confidence_level, df, loc=x_bar, scale=sem) calculates a confidence interval for the mean of a population, assuming the data follows a t-distribution. Here's a step-by-step breakdown:

confidence_level: This is the desired confidence level for the interval (e.g., 0.95 for a 95% CI).
df: This stands for 'degrees of freedom', which is n - 1 (where n is the sample size). It determines the shape of the t-distribution.
loc=x_bar: This sets the center of the interval to the sample mean (x̄).
scale=sem: This provides the standard error of the mean (SEM), which is s / sqrt(n), where s is the sample standard deviation and n is the sample size. The scale parameter essentially stretches or shrinks the interval based on the variability of the sample means.
Internally, the function calculates the appropriate t-critical values (t*) for the given confidence_level and df, and then computes the interval as x_bar ± t* * sem.

In other words, the sample mean is not statistically significantly different from the population mean.
If a sample's CI does NOT contain the population mean (red lines in the plot, though none appeared in this specific run): This means that based on this sample, we reject the null hypothesis. The sample mean is statistically significantly different from the population mean.
If we were comparing two independent sample means (e.g., treatment vs. control) and their respective CIs overlap, it generally suggests that there is no statistically significant difference between the true population means of the two groups at the chosen confidence level. If the CIs do not overlap, it suggests a significant difference. It's important to note that direct overlap of CIs is a conservative test; non-overlap is always significant, but overlap does not always mean non-significance.

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">What Is a Sampling Distribution?</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">A <strong>sampling distribution</strong> is the probability distribution of a statistic (e.g., x̄, p̂) obtained from <em>all possible</em> samples of size n drawn from a population.</p>
  <ul style="color:#333; line-height:1.8;">
    <li><strong>Step 1:</strong> Start with a population described by μ and σ.</li>
    <li><strong>Step 2:</strong> Draw many random samples (S₁, S₂, …, Sₖ) of the same size n.</li>
    <li><strong>Step 3:</strong> Compute the statistic for each sample (x̄₁, x̄₂, …, x̄ₖ) and plot the distribution.</li>
  </ul>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">The resulting distribution is centered at μ with spread σ/√n. This is the foundation of every confidence interval and hypothesis test you will ever compute.</p>
</div>

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">Sampling Variability &amp; Bias</h2>
</div>

**Sampling variability** is the natural variation in a statistic from sample to sample. It is *not* an error — it is an inherent property of sampling.

| Sources | Reducing Variability |
|:--|:--|
| Random selection — different individuals chosen | Increase *n* — SE = σ/√n shrinks |
| Population heterogeneity | Better sampling (stratification, clustering) |
| Small sample size | More precise measurements |

**Bias.** An estimator is *unbiased* if E(estimator) = parameter. Examples:
- E(x̄) = μ ✓
- E(p̂) = π ✓
- E(s²) = σ² (sample variance with df = n − 1) ✓

But using n (instead of n − 1) for the sample variance produces a **biased** estimator: E[Σ(xᵢ − x̄)² / n] = (n − 1)/n · σ² < σ². Non-random sampling (convenience, voluntary response, survivorship) also introduces bias.

> **MSE = Bias² + Variance.** A good estimator is both unbiased (centred on the target) and has low variance (tightly clustered).

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">History &amp; Intuition of the CLT</h2>
</div>

| Year | Contributor | Contribution |
|:----:|:------------|:-------------|
| 1733 | Abraham de Moivre | First observed that binomial distributions approximate the normal curve. |
| 1812 | Pierre-Simon Laplace | Generalised to sums of independent random variables. |
| 1920 | Lindeberg–Lévy | Proved the modern CLT under general conditions (i.i.d. with finite variance). |

**Modern formulation.** If X₁, …, Xₙ are i.i.d. with mean μ and finite variance σ², then:

$$\sqrt{n}\;\frac{\bar X - \mu}{\sigma} \;\longrightarrow\; N(0, 1) \quad\text{as } n \to \infty$$

**Three intuitions:**
1. **Cancellation of extremes** — averaging many values lets highs and lows cancel.
2. **Accumulation of information** — each observation adds a small piece; outliers lose influence.
3. **Probability concentration** — by the Law of Large Numbers, x̄ → μ; the CLT tells us the *shape* (normal) and *rate* (σ/√n) of that convergence.

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

mu, sigma, n, reps = 100, 15, 10, 10000
true_var = sigma ** 2

biased_variances = []
unbiased_variances = []

for _ in range(reps):
    sample = np.random.normal(loc=mu, scale=sigma, size=n)
    biased_variances.append(np.var(sample, ddof=0))
    unbiased_variances.append(np.var(sample, ddof=1))

biased_variances = np.array(biased_variances)
unbiased_variances = np.array(unbiased_variances)

fig, axes = plt.subplots(1, 2, figsize=(14, 6), sharey=True)
fig.suptitle('Sampling Distribution of Biased vs. Unbiased Variance Estimators', fontsize=16, color='#4B2E83', fontweight='bold')

# Plot for Biased Variance
axes[0].hist(biased_variances, bins=50, density=True, alpha=0.7, color='#4B2E83', edgecolor='black')
axes[0].axvline(true_var, color='red', linestyle='--', linewidth=2, label=f'True $\sigma^2$ = {true_var}')
axes[0].set_title('Biased Variance (ddof=0)', fontsize=12, color='#4B2E83')
axes[0].set_xlabel('Variance')
axes[0].set_ylabel('Density')
axes[0].legend()
axes[0].grid(axis='y', alpha=0.75)

# Plot for Unbiased Variance
axes[1].hist(unbiased_variances, bins=50, density=True, alpha=0.7, color='#E8B923', edgecolor='black')
axes[1].axvline(true_var, color='red', linestyle='--', linewidth=2, label=f'True $\sigma^2$ = {true_var}')
axes[1].set_title('Unbiased Variance (ddof=1)', fontsize=12, color='#4B2E83')
axes[1].set_xlabel('Variance')
axes[1].legend()
axes[1].grid(axis='y', alpha=0.75)

plt.tight_layout(rect=[0, 0.03, 1, 0.9])
plt.show()

print(f'True Population Variance (σ²): {true_var:.3f}')
print(f'Mean of Biased Variances (E[S²_biased]): {np.mean(biased_variances):.3f}')
print(f'Mean of Unbiased Variances (E[S²_unbiased]): {np.mean(unbiased_variances):.3f}')
````

Why does ddof=0 underestimate σ²? When ddof=0 is used for variance calculation (dividing by n instead of n-1), the formula computes the population variance assuming the sample itself is the population. However, when we are using a sample to estimate the variance of a larger population, the sample's variability is typically smaller than the population's true variability because the sample usually doesn't capture the full range of extreme values present in the population. The sample mean (x̄) is also used in the calculation, and x̄ is always closer to the sample's own data points than the true population mean (μ) would be. This makes the sum of squared differences Σ(xᵢ - x̄)² smaller on average than Σ(xᵢ - μ)². Therefore, dividing by n (which is larger than n-1) compounds this underestimation, leading to a biased estimator that consistently gives values lower than the true population variance.

Which histogram centers on 225? The unbiased variance (ddof=1) histogram is centered on 225, which is the true population variance (σ² = 15² = 225). Looking at the printed means: Mean of Unbiased Variances (E[S²_unbiased]): 223.444 is very close to 225. In contrast, the Mean of Biased Variances (E[S²_biased]): 201.099 is noticeably lower than 225.

What does "unbiased" mean in terms of E(s²)? In statistics, an estimator is considered unbiased if its expected value (the average value you would get if you took an infinite number of samples and calculated the estimator for each) is equal to the true population parameter it is trying to estimate. For the sample variance, using ddof=1 (dividing by n-1, known as Bessel's correction) provides an unbiased estimator. This means that on average, the sample variance calculated with ddof=1 will correctly estimate the true population variance (σ²). Our simulation results support this: the mean of the unbiased variances (223.444) is very close to the true population variance (225), demonstrating its unbiased nature over many samples.

**Standard Error (SE)** — the standard deviation of a *sampling distribution*. It quantifies how much a statistic varies from sample to sample.

| Statistic | Formula |
|:--|:--|
| Sample mean | SE = σ / √n |
| Sample proportion | SE = √( p̂(1 − p̂) / n ) |

Larger n → smaller SE → more precise estimates.

**Confidence Interval (CI)** — a range of values likely to contain the true parameter with a specified confidence level.

$$\text{95\% CI:}\quad \text{estimate} \;\pm\; z^* \cdot \text{SE}$$

**Interpretation:** if we repeated the sampling many times, 95% of the CIs would contain the true parameter.

> Error bars on a chart are **visual confidence intervals**. If the error bar for a sample proportion includes the hypothesised value, we cannot reject H₀ at that significance level — directly linking CIs to hypothesis testing.

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">Z vs Student's t</h2>
</div>

| | Normal (Z) | Student's t |
|:--|:--|:--|
| Formula | Z = (x̄ − μ) / (σ/√n) | t = (x̄ − μ) / (*s*/√n) |
| Use when | σ known, n ≥ 30 (CLT) | σ unknown, small n |
| Shape | Fixed symmetric bell curve | Heavier tails, depends on df = n − 1 |
| 95% critical value | z* = 1.96 | t* depends on df |

**95% critical values (two-sided):**

| df | 2 | 5 | 10 | 20 | 30 | 120 | ∞ (Z) |
|:--:|:-:|:-:|:-:|:-:|:-:|:-:|:-:|
| t* | 4.303 | 2.571 | 2.228 | 2.086 | 2.042 | 1.980 | **1.960** |

As df → ∞, t* → z*. For large samples the two distributions are indistinguishable.

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">Hypothesis Testing</h2>
</div>

| Symbol | Name | Role |
|:--|:--|:--|
| **H₀** | Null hypothesis | Default claim (always contains `=`). Assumed true until the data say otherwise. |
| **H₁** | Alternative hypothesis | The research claim (contains `≠`, `<`, `>`). Accepted only if H₀ is rejected. |
| **α** | Significance level | Probability of rejecting H₀ when it is actually true (Type I error). Typically 0.05. |
| **p-value** | | Probability of observing data as extreme as ours, assuming H₀ is true. |

**Decision rule:**

- If p-value < α → **reject H₀**
- If p-value ≥ α → **fail to reject H₀**

**CI ↔ Hypothesis test duality.** A (1 − α) confidence interval contains *all* values of the null hypothesis that would **not** be rejected by a two-sided test at level α:

- If μ₀ ∈ CI → fail to reject H₀
- If μ₀ ∉ CI → reject H₀

A 95% CI is therefore equivalent to a two-sided hypothesis test at α = 0.05.

````python
mu, sigma, n, reps = 50, 10, 25, 100
z_star = 1.96

# --- Your code here ---

sem = sigma / np.sqrt(n) # Standard Error of the Mean

fig, ax = plt.subplots(figsize=(10, 8))
ax.axvline(mu, color='red', linestyle='--', label=f'True Mean (μ={mu})')

coverage_count = 0
for i in range(reps):
    sample = np.random.normal(loc=mu, scale=sigma, size=n)
    x_bar = np.mean(sample)

    # Calculate 95% CI
    ci_lower = x_bar - z_star * sem
    ci_upper = x_bar + z_star * sem

    # Check if CI contains mu
    if ci_lower <= mu <= ci_upper:
        color = 'green'
        coverage_count += 1
    else:
        color = 'red'

    # Plot CI as a horizontal line segment
    ax.hlines(y=i, xmin=ci_lower, xmax=ci_upper, color=color, linewidth=2)
    ax.plot(x_bar, i, 'o', color=color, markersize=6) # Plot sample mean as a dot

ax.set_yticks([])
ax.set_xlabel('Value')
ax.set_ylabel('Sample Index')
ax.set_title(f'95% Confidence Intervals for {reps} Samples from N({mu}, {sigma})')
ax.legend()
plt.tight_layout()
plt.show()

coverage_percentage = (coverage_count / reps) * 100
print(f'Number of CIs that captured μ: {coverage_count} out of {reps} ({coverage_percentage:.2f}%)')
print(f'Standard Error (SE = σ/√n): {sem:.3f}')
````

2. **SE value** — The Standard Error (SE = σ/√n) in this scenario is **2.000**. This is calculated using the population standard deviation (σ=10) and the sample size (n=25): SE = 10 / √25 = 10 / 5 = 2.0.

3. **Effect of 99% CIs** — If you used 99% CIs instead of 95% CIs, the plot would change in a few key ways:
    *   **Wider Intervals:** Each horizontal line segment representing a CI would be wider. This is because a higher confidence level (99% vs 95%) requires a larger critical value (z*). For a 99% CI, z* would be approximately 2.576 instead of 1.96, leading to a larger margin of error (z* * SE).
    *   **Higher Coverage:** A larger proportion of the CIs (closer to 99%) would be green, indicating they captured the true mean. This is the purpose of a higher confidence level – to increase the probability that the interval contains the true parameter.
    *   **Fewer Red Lines:** Consequently, there would be fewer red lines, as fewer CIs would fail to capture the true mean.

💬 **LLM Prompt**

```
Write Python code that draws 5,000 samples of size n=5 from N(50, 10). For each sample compute:
1) A 95% CI using Z (z*=1.96) with sample s
2) A 95% CI using t (df=n-1) with sample s
Plot side-by-side: left=Z CIs, right=t CIs (show first 50 samples). Color green if CI contains mu=50,
red if not. Print coverage % for both methods.
Also print t*(df=4) vs z*=1.96 and average CI width.
```

🎯 **Task:** Run the code. Explain: (1) Which method has coverage closer to 95%? (2) Why are t-based CIs wider? (3) What is t*(df=4) vs z*? (4) When does the difference vanish?

````python
mu, sigma, n, reps = 50, 10, 5, 5000
z_star = 1.96
t_star = stats.t.ppf(0.975, df=n-1)

# Lists to store CI results
z_cis = []
t_cis = []
z_coverage = []
t_coverage = []

for _ in range(reps):
    sample = np.random.normal(loc=mu, scale=sigma, size=n)
    x_bar = np.mean(sample)
    s = np.std(sample, ddof=1) # Sample standard deviation (unbiased)
    sem = s / np.sqrt(n)       # Standard error of the mean (using sample std dev)

    # 1) Compute 95% CI using Z (z*=1.96) with sample s
    z_ci_lower = x_bar - z_star * sem
    z_ci_upper = x_bar + z_star * sem
    z_cis.append((z_ci_lower, z_ci_upper))
    z_coverage.append(z_ci_lower <= mu <= z_ci_upper)

    # 2) Compute 95% CI using t (df=n-1) with sample s
    t_ci_lower = x_bar - t_star * sem
    t_ci_upper = x_bar + t_star * sem
    t_cis.append((t_ci_lower, t_ci_upper))
    t_coverage.append(t_ci_lower <= mu <= t_ci_upper)

# Plotting the first 50 samples
num_to_plot = 50

fig, axes = plt.subplots(1, 2, figsize=(14, 8), sharey=True)
fig.suptitle(f'95% Confidence Intervals: Z vs t-distribution (n={n}, first {num_to_plot} samples)', fontsize=16, color='#4B2E83', fontweight='bold')

# Plot for Z CIs
axes[0].axvline(mu, color='red', linestyle='--', label=f'True Mean (μ={mu})')
for i in range(num_to_plot):
    color = 'green' if z_coverage[i] else 'red'
    axes[0].hlines(y=i, xmin=z_cis[i][0], xmax=z_cis[i][1], color=color, linewidth=2)
    axes[0].plot(np.mean(np.random.normal(loc=mu, scale=sigma, size=n)), i, 'o', color=color, markersize=6)
axes[0].set_title('Z-distribution CIs (using s)', fontsize=12, color='#4B2E83')
axes[0].set_xlabel('Value')
axes[0].set_ylabel('Sample Index')
axes[0].set_yticks([])
axes[0].legend()

# Plot for t CIs
axes[1].axvline(mu, color='red', linestyle='--', label=f'True Mean (μ={mu})')
for i in range(num_to_plot):
    color = 'green' if t_coverage[i] else 'red'
    axes[1].hlines(y=i, xmin=t_cis[i][0], xmax=t_cis[i][1], color=color, linewidth=2)
    axes[1].plot(np.mean(np.random.normal(loc=mu, scale=sigma, size=n)), i, 'o', color=color, markersize=6)
axes[1].set_title('t-distribution CIs (using s)', fontsize=12, color='#4B2E83')
axes[1].set_xlabel('Value')
axes[1].set_yticks([])
axes[1].legend()

plt.tight_layout(rect=[0, 0.03, 1, 0.9])
plt.show()

# Print coverage percentages
z_coverage_percent = (np.sum(z_coverage) / reps) * 100
t_coverage_percent = (np.sum(t_coverage) / reps) * 100
print(f'\nZ-CI Coverage: {z_coverage_percent:.2f}% of {reps} samples captured μ')
print(f't-CI Coverage: {t_coverage_percent:.2f}% of {reps} samples captured μ')

# Print t*(df=4) vs z*=1.96
print(f'\nCritical Values (95% CI, two-sided):')
print(f'  Z* = {z_star:.3f}')
print(f'  t*(df={n-1}) = {t_star:.3f}')

# Calculate average CI width
avg_z_width = np.mean([upper - lower for lower, upper in z_cis])
avg_t_width = np.mean([upper - lower for lower, upper in t_cis])
print(f'\nAverage CI Width:')
print(f'  Z-CI Average Width: {avg_z_width:.3f}')
print(f'  t-CI Average Width: {avg_t_width:.3f}')
````

*Write your written answer here.*

1.  **Which method has coverage closer to 95%?** — The simulation results clearly show that the **t-CI Coverage** is much closer to the target 95% (e.g., 94.62%) compared to the Z-CI Coverage (e.g., 85.58%). This is expected because the t-distribution correctly accounts for the additional uncertainty when the population standard deviation (`σ`) is unknown and estimated by the sample standard deviation (`s`) from a small sample. The Z-CI, by using `z_star` but `s` instead of `σ`, effectively underestimates the variability, leading to CIs that are too narrow and thus capture the true mean less often.

2.  **Why are t-based CIs wider?** — The t-based CIs are wider than the Z-based CIs because the critical value for the t-distribution (`t*`) is larger than the critical value for the Z-distribution (`z*`) for small sample sizes. Specifically, for `n=5` (and thus `df=4`), `t*` is approximately 2.776, whereas `z*` is 1.960. A larger critical value, when multiplied by the standard error, results in a larger margin of error and thus a wider confidence interval. This wider interval reflects the increased uncertainty associated with estimating the population standard deviation from a small sample.

3.  **What is t*(df=4) vs z*?** —
    *   `z*` (for a 95% two-sided CI) is **1.960**. This value comes from the standard normal distribution and is used when the population standard deviation `σ` is known or when the sample size is very large.
    *   `t*(df=4)` (for a 95% two-sided CI with 4 degrees of freedom, which corresponds to `n-1 = 5-1`) is approximately **2.776**. This value comes from the t-distribution and is larger than `z*` to account for the increased variability introduced by using the sample standard deviation `s` as an estimate for `σ` with a small sample.

4.  **When does the difference vanish?** — The difference between the t-distribution and the Z-distribution (and thus between `t*` and `z*`) vanishes as the degrees of freedom (`df`) approach infinity, which in practice means as the sample size (`n`) becomes very large (typically `n >= 30` or `n >= 60`). As `n` increases, the sample standard deviation `s` becomes a very good estimate of the population standard deviation `σ`, and the t-distribution converges to the Z-distribution. Consequently, `t*` values become very close to `z*` values, and the t-based CIs become practically indistinguishable from Z-based CIs.

````python
# --- Your code here ---
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import numpy as np # numpy is needed for np.nan
import pandas as pd # pandas is needed for DataFrame

# Corrected: Load the CSV from the GitHub URL within this cell
url = 'https://raw.githubusercontent.com/PedroBSB/BDATA200/main/Data/winequality-red.csv'
wine = pd.read_csv(url)

# Define the confidence level
confidence_level = 0.95

# Group by quality and calculate stats for alcohol and volatile_acidity
wine_grouped = wine.groupby('quality')[['alcohol', 'volatile_acidity']].agg(['mean', 'sem', 'count'])

# Flatten the multi-level columns
wine_grouped.columns = ['_'.join(col).strip() for col in wine_grouped.columns.values]

# Calculate the 95% CI for each mean
def calculate_ci(row, feature):
    mean = row[f'{feature}_mean']
    sem = row[f'{feature}_sem']
    count = row[f'{feature}_count']

    # Degrees of freedom
    df = count - 1

    # If degrees of freedom is 0, t-distribution is not defined, return nan or handle appropriately
    if df <= 0:
        return np.nan, np.nan

    # Calculate t-critical value
    t_crit = stats.t.ppf(1 - (1 - confidence_level) / 2, df)

    margin_of_error = t_crit * sem
    return mean - margin_of_error, mean + margin_of_error

wine_grouped['alcohol_ci_lower'], wine_grouped['alcohol_ci_upper'] = zip(*wine_grouped.apply(lambda row: calculate_ci(row, 'alcohol'), axis=1))
wine_grouped['va_ci_lower'], wine_grouped['va_ci_upper'] = zip(*wine_grouped.apply(lambda row: calculate_ci(row, 'volatile_acidity'), axis=1))

# Prepare data for plotting
plot_data = []
for quality in wine_grouped.index:
    plot_data.append({
        'quality': quality,
        'feature': 'alcohol',
        'mean': wine_grouped.loc[quality, 'alcohol_mean'],
        'ci_lower': wine_grouped.loc[quality, 'alcohol_ci_lower'],
        'ci_upper': wine_grouped.loc[quality, 'alcohol_ci_upper']
    })
    plot_data.append({
        'quality': quality,
        'feature': 'volatile_acidity',
        'mean': wine_grouped.loc[quality, 'va_mean'],
        'ci_lower': wine_grouped.loc[quality, 'va_ci_lower'],
        'ci_upper': wine_grouped.loc[quality, 'va_ci_upper']
    })
plot_df = pd.DataFrame(plot_data)

# Create the grouped bar chart with error bars
plt.figure(figsize=(12, 7))
sns.barplot(
    x='quality',
    y='mean',
    hue='feature',
    data=plot_df,
    palette={'alcohol': '#4B2E83', 'volatile_acidity': '#E8B923'},
    errorbar=None, # We'll draw custom error bars
    edgecolor='black'
)

# Add custom error bars
for i, row in plot_df.iterrows():
    if not pd.isna(row['ci_lower']) and not pd.isna(row['ci_upper']): # Only plot if CI is valid
        # Adjust x position for grouped bars using the seaborn barplot's internal positions
        # This requires re-calculating or finding the bar positions. A simpler way for fixed groups is to offset.
        # For exact positioning, one would get bar positions after sns.barplot renders.
        # For this exercise, an approximation is often sufficient for visual grouping.
        # Let's adjust based on the order and hue levels
        if row['feature'] == 'alcohol':
            # Alcohol bars are usually on the left in each group
            x_offset = -0.2
        else: # volatile_acidity
            # Volatile acidity bars are usually on the right in each group
            x_offset = 0.2

        # Find the x-coordinate of the center of the bar for the current quality and feature
        # This is a bit tricky with sns.barplot not returning bar objects directly for custom error bars
        # A common workaround is to use the quality value directly as x, and the offset for grouping
        plt.errorbar(
            x=row['quality'] + x_offset,
            y=row['mean'],
            yerr=[[row['mean'] - row['ci_lower']], [row['ci_upper'] - row['mean']]],
            fmt='none',  # No line connecting points
            capsize=5,  # Size of the caps on the error bars
            color='black', # Color of the error bars
            elinewidth=1.5 # Width of the error bar lines
        )

plt.title('Mean Alcohol and Volatile Acidity by Wine Quality with 95% CI', fontsize=16, color='#4B2E83', fontweight='bold')
plt.xlabel('Quality Score', fontsize=12, color='#333')
plt.ylabel('Mean Value', fontsize=12, color='#333')
plt.legend(title='Feature')
plt.grid(axis='y', alpha=0.7)
plt.xticks(sorted(wine['quality'].unique()))
plt.tight_layout()
plt.show()
````
